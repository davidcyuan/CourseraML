```python
def zscore_normalize(X):
    # X is matrix of training data, ndarray
    X_mean = np.mean(X, axis=0)
    X_std = np.std(X, axis=0)
    X_norm = (X - X_mean) / X_std

    return (X_norm, X_mean, X_std)
```

```python
X_train_norm, X_train_mean, X_train_std = zscore_normalize(X_train)
```

```python
def cost_function(X, y, w, b):
    # X is training matrix, m examples of n features : ndarray
    # y is target, m examples : ndarray
    # w is feature weights, n weights : ndarray
    # b is constant : float

    example_num = X.shape[0]
    Y_predictions = np.matmul(X, w) + b
    Y_errors = np.subtract(Y_predictions, y)
    cost = np.square(Y_errors)
    cost = np.sum(cost) / 2 / example_num
    
    return cost
```

```python
def gradient(X, y, w, b):
    # X is training matrix, m examples of n features : ndarray
    # y is target, m examples : ndarray
    # w is feature weights, n weights : ndarray
    # b is constant : float

    # returns gradients of features as ndarray, and gradient of b as float

    example_num = X.shape[0]
    Y_predictions = np.matmul(X, w) + b
    Y_errors = np.subtract(Y_predictions, y)

    # Transpose X to matrix multiple with Y_errors, getting sum of errors for every feature across all examples
    X_transpose = X.transpose()
    errors_times_features = np.matmul(X_transpose, Y_errors)
    gradient_features = errors_times_features / example_num
    gradient_b = np.sum(Y_errors) / example_num

    return gradient_features, gradient_b
    
    
```

```python
def gradient_descent(X_train, y, w_init, b_init, alpha, iters):
    # X_train : training data,                      (m, n) ndarray
    # y       : target values,                      (m,  ) ndarray
    # w_init  : initial feature weights,            (n,  ) ndarray
    # b_init  : initial constant,                          float
    # alpha   : learning rate                              float
    # iters   : number of iterations                       int
        #       initial state counts as 1 iteration

    m, n = X_train.shape
    
    J_history = np.zeros(iters)
        # Cost history
    w_history = np.zeros((iters, n))
        # feature weights history
    b_history = np.zeros(iters)
        # constant history
    
    w_current = w_init
    b_current = b_init
    
    for iteration in range(iters):
        J_history[iteration] = cost_function(X_train, y, w_current, b_current)
        w_history[iteration] = w_current
        b_history[iteration] = b_current
        if (iteration % 100 == 0) or iteration >= (iters - 10) :
            print(f"Iteration: {iteration}    Cost: {J_history[iteration]}")

        # get gradients
        g_w, g_b = gradient(X_train, y, w_current, b_current)

        w_current = np.subtract(w_current, g_w * alpha)
        b_current = b_current - g_b * alpha

    return J_history, w_history, b_history, 
```
