```python
from sklearn import datasets
import numpy as np
np.set_printoptions(precision=3)
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler

from MyLRFunctions import print_scatter
```

```python
housing_bunch = datasets.fetch_california_housing()

X_train = housing_bunch["data"]
Y_train = housing_bunch["target"]
feature_names = housing_bunch["feature_names"]
target_name = "Median House Value (100k's)"

#trim data for easier learning
X_train = X_train[0:500,:]
Y_train = Y_train[0:500]
```

```python
scaler = StandardScaler()
X_norm = scaler.fit_transform(X_train)
```

```python
sgdr = SGDRegressor(max_iter = 1000)
sgdr.fit(X_norm, Y_train)
```

```python
b_norm = sgdr.intercept_
w_norm = sgdr.coef_
print(f"w_norm: {w_norm}\n b_norm: {b_norm}")
```

```python
y_predict = sgdr.predict(X_norm)
print(f"{y_predict[:10]}")
print(Y_train[:10])
```

```python
print_scatter(X_train, Y_train, feature_names, y_predict)
```
